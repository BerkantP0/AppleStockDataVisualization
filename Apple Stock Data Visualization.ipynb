# 1. Libraries and Data Retrieval

```python
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
```

# 2. Data Retrieval and Basic Visualization

```python
ticker = 'AAPL'  # Apple Inc.
data = yf.download(ticker, start='2020-01-01', end='2024-01-01')

plt.figure(figsize=(12, 6))
sns.lineplot(data=data['Close'], label='Closing Price')
plt.title(f'{ticker} Stock Closing Prices (2020-2024)')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

# 3. Calculating and Visualizing Moving Averages

```python
data['MA50'] = data['Close'].rolling(window=50).mean()
data['MA200'] = data['Close'].rolling(window=200).mean()

plt.figure(figsize=(12, 8))
plt.plot(data['Close'], label='Closing Price')
plt.plot(data['MA50'], label='50-Day Moving Average')
plt.plot(data['MA200'], label='200-Day Moving Average')
plt.title(f'{ticker} Stock Prices and Moving Averages (2020-2023)')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

# 4. Volume Visualization

```python
plt.figure(figsize=(12, 6))
plt.bar(data.index, data['Volume'], color='grey', label='Volume')
plt.title(f'{ticker} Stock Volume (2020-2023)')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()
plt.show()
```
